```python
# dataset for the further analysis
analysis_data = data[['ResponseId',
                      'Age',
                      'Country',
                      'Employment',
                      'ConvertedCompYearly',
                      'OrgSize',
                      'DevType']]
analysis_data.shape
```

```python
# Do you need to worry your profession after 35 year old in the IT industry?
# metric - a. unemployment rate
# metric - b. salary (mean/median/variance)

# Q1. by region
# Q2. by OrgSize
# Q3. by DevType
```

```python
# remove rows that Age is null
analysis_data = analysis_data.dropna(subset=['Age'])
```

```python
analysis_data.shape
```

```python
analysis_data.groupby(['Age']).nunique()
```

```python
analysis_data.head()
```

```python
# group groups with smaller size into one large group
analysis_data['Age'] = analysis_data['Age'].replace('45-54 years old', '45 years old and beyond')
analysis_data['Age'] = analysis_data['Age'].replace('55-64 years old', '45 years old and beyond')
analysis_data['Age'] = analysis_data['Age'].replace('65 years or older', '45 years old and beyond')

# remove rows that Age = Prefer not to say
analysis_data = analysis_data.loc[analysis_data['Age'] != 'Prefer not to say']
```

```python
# sort age by a order from yound to old
analysis_data["Age"] = pd.Categorical(analysis_data["Age"], categories = ['Under 18 years old', 
                                         '18-24 years old', 
                                        '25-34 years old',
                                        '35-44 years old',
                                        '45 years old and beyond'])
```

```python
analysis_data.groupby(['Age']).nunique()
```

```python
analysis_data.groupby(['Employment']).nunique()
```

```python
# group groups with smaller size into one large group
analysis_data['Employment'] = analysis_data['Employment'].replace('Employed full-time', 'Employed')
analysis_data['Employment'] = analysis_data['Employment'].replace('Employed part-time', 'Employed')
analysis_data['Employment'] = analysis_data['Employment'].replace('Independent contractor, freelancer, or self-employed', 'Employed')
analysis_data['Employment'] = analysis_data['Employment'].replace('Employed part-time', 'Employed')
analysis_data['Employment'] = analysis_data['Employment'].replace('Not employed, and not looking for work', 'Unemployed')
analysis_data['Employment'] = analysis_data['Employment'].replace('Not employed, but looking for work', 'Unemployed')
analysis_data['Employment'] = analysis_data['Employment'].replace('Student, full-time', 'Student')
analysis_data['Employment'] = analysis_data['Employment'].replace('Student, part-time', 'Student')


# remove rows that Age = Prefer not to say
analysis_data = analysis_data.loc[analysis_data['Employment'] != 'I prefer not to say']
```

```python
analysis_data.groupby(['Employment']).nunique()
```

```python
# sort Employment 
analysis_data["Employment"] = pd.Categorical(analysis_data["Employment"], categories = ['Student', 
                                         'Unemployed', 
                                        'Retired',
                                        'Employed'])
```

```python
# aggregated by age and employment status
age_employ_aggregated = analysis_data.groupby(['Age','Employment']).nunique().reset_index()[['Age','Employment','ResponseId']]
```

```python
# total number of participants in each age group
age_employ_aggregated['sum']  = age_employ_aggregated['ResponseId'].groupby(age_employ_aggregated['Age']).transform('sum')
```

```python
# % of participants employed by each age group
age_employ_aggregated['share'] = round(age_employ_aggregated['ResponseId'] * 100 / age_employ_aggregated['sum'],1)
```

```python
## 35-44 actually has the highest employment rate
## however, the unemployed % increase as participant group from 35-44 to 45 and beyond.
## global overall
age_employ_aggregated.loc[age_employ_aggregated['Employment'].isin(['Employed','Retired'])]
```

```python
# find top 50 countries regarding number of participants
top_countries = list(analysis_data.groupby(['Country']).nunique().reset_index().sort_values(by='ResponseId',ascending = False).head(50)['Country'])
```

```python
analysis_data_top_country = analysis_data.loc[analysis_data['Country'].isin(top_countries)]
```

```python
analysis_data_top_country
```

```python
analysis_data_top_country_aggr = analysis_data_top_country.groupby(['Age','Country','Employment']).nunique().reset_index()[['Age','Country','Employment','ResponseId']]
analysis_data_top_country_aggr
```

```python
analysis_data_top_country_aggr['sum'] = analysis_data_top_country_aggr.groupby(['Age','Country']).transform('sum')

analysis_data_top_country_aggr['share'] = round(analysis_data_top_country_aggr['ResponseId'] * 100 / analysis_data_top_country_aggr['sum'],1)

unemployed_rate_country = analysis_data_top_country_aggr.loc[analysis_data_top_country_aggr['Employment'] == 'Unemployed']
```
